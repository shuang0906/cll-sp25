```python
import random
import math
from IPython.display import display, HTML
def show_html(src):
    return display(HTML(src), metadata=dict(isolated=True))
def mkdiv(content, **kwargs):
    if 'position' not in kwargs:
        kwargs['position'] = 'absolute'
    style_str = ' '.join([": ".join((k.replace('_', '-'), v))+";" for k, v in kwargs.items()])
    return f"<div style='{style_str}'>{content}</div>"
```

```python
html_tmpl = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="utf-8">
    <title>{title}</title>
    <style>
    html {{ min-height: 32em; overflow: hidden; }}
    div {{ color:black; background-color: white; outline: 1px solid black; width: auto;height: auto; aspect-ratio: 1 / 1;}}
    </style>
</head>
<body>
{content}
</body>
</html>"""

src = open("frost.txt").read()
letters = list(src)
grid_size = int(math.sqrt(len(letters)))
current_index = 0
divs = []
for i in range(grid_size):
    for j in range(grid_size):
        x_pos = 5 + (90 / grid_size) * i
        y_pos = 5 + (90 / (grid_size)) * j
        this_div = mkdiv(letters[current_index],
                         top=f"{x_pos}%",
                         left=f"{y_pos}%",
                         font_family="Helvetica",
                         font_size=f"{random.randrange(4, 48)}",
                         transform="translate(-50%, -50%)")
        divs.append(this_div)
        current_index += 1
html_src = html_tmpl.format(title="frost grid", content="".join(divs))
show_html(html_src)
open("output.html" , "w", encoding="utf-8"). write(html_src)
```

```python
import random
import math

def mkdiv(content, **kwargs):
    style_str = ' '.join([": ".join((k.replace('_', '-'), v)) + ";" for k, v in kwargs.items()])
    return f"<div style='{style_str}'>{content}</div>"

def mksec(content, **kwargs):
    style_str = ' '.join([": ".join((k.replace('_', '-'), v)) + ";" for k, v in kwargs.items()])
    return f"<section style='{style_str}'>{content}</section>"

# HTML Template
html_tmpl = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="utf-8">
    <title>{title}</title>
    <style>
    html {{ overflow-x: hidden;}}
    body {{ margin: 0px;}}
    section {{ position: absolute; display: grid; grid-auto-flow: row dense; overflow: hidden;}}
    div {{ color: gray; display: flex; justify-content: center; align-items: center; }}
    div:hover {{filter: blur(10px);}}
    </style>
</head>
<body>
    {content}
</body>
</html>"""

color_list = ["#fd61eb", "#5eaf66", "#dfff47", "#4c56f5", "#000000", "#b26dfe", "#FFFFFF", "red"]

# Read input letters and compute grid size
src = open("frost.txt").read()
letters = list(src)
grid_size = int(math.sqrt(len(letters)))

# Generate divs
all_divs = []
for i in range(grid_size):
    for j in range(grid_size):
        span_size = random.randint(1, 3)
        current_index = i * grid_size + j
        if current_index >= len(letters):
            break
        div_style = {
            "font_family": "Helvetica",
            "font_weight": "600",
            "font_size": f"{span_size*6}px",
            "color": f"rgb({random.randint(0,255)}, {random.randint(0,255)}, {random.randint(0,255)})",
            "grid_column_end": f"span {span_size}",
            "grid_row_end": f"span {span_size}",
            "background_color": random.choice(color_list),
            #"border": f"4px solid black"
        }
        #all_divs.append(mkdiv("E", **div_style))
        all_divs.append(mkdiv(letters[current_index], **div_style))
        
# Generate sections
sections = []

# Add a full-screen section
fullscreen_section_style = {
    "width": "100vw",
    "height": "calc(100vw/16*9)",
    "left": "0",
    "top": "0",
    "grid_template_columns": f"repeat(100, 1vw)",
    "grid_template_rows": f"repeat(calc(100/16*9), 1vw)",
    "filter": "blur(4px)"
}

sections.append(mksec("".join(all_divs), **fullscreen_section_style))

for _ in range(3):
    w_size = random.randint(10,40)
    h_size = random.randint(10,40)
    den_size = random.randint(1,3)
    section_style = {
        #"background_color": f"rgb({random.randint(0,255)}, {random.randint(0,255)}, {random.randint(0,255)})",
        "width": f"{w_size}vw",
        "height": f"{h_size}vw",
        "left": f"{random.randint(0,100-w_size)}vw",
        "top": f"{random.randint(0, int((100/16*9) - h_size))}vw",
        "grid_template_columns": f"repeat({w_size*den_size}, {1/den_size}vw)",
        "grid_template_rows": f"repeat({h_size*den_size}, {1/den_size}vw)"
    }
    sections.append(mksec("".join(all_divs), **section_style))

# Generate HTML
html_src = html_tmpl.format(title="frost grid", content="".join(sections))

# Write to an output HTML file
with open("output.html", "w", encoding="utf-8") as f:
    f.write(html_src)

print("HTML file with styled sections generated successfully.")
#DM Mono, Copper
```
